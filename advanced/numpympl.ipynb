**Problem:** Try to swap parts of the image.

If you split the image into 4 parts like:

AB
CD

it should become:

AC
BD

```python
face2 = face.copy()
```

```python
h,w = face2.shape
```

```python
face2 = face.copy()
```

```python
h, w = face2.shape
```

```python
TR = face2[:h//2,w//2:].copy()
BL = face2[h//2:,:w//2].copy()
imshow(face2)
imshow(TR)
imshow(BL)
```

```python
face2[:h//2, w//2:] = BL
face2[h//2:, :w//2] = TR
imshow(face2)
```

### Example: temperature data ###

Download the data from [HYDERABAD-weather.csv](https://notes.pipal.in/2017/arcesium-oct-advpython/HYDERABAD-weather.csv)

```python
import csv
```

```python
data = list(csv.reader(open("HYDERABAD-weather.csv")))
```

```python
data[:3]
```

```python
data = data[1:] #skip the header
```

```python
tmax = [float(row[4]) for row in data]
tmin = [float(row[5]) for row in data]
```

```python
plt.scatter(tmin, tmax)
plt.show()
```

**Problem:** Plot a scatter plot between min temperature and rain fall.

```python
rainfall = [float(row[6]) for row in data]
```

```python
def safefloat(value):
    try:
        return float(value)
    except ValueError:
        print("bad value: %r" % value)
        return 0.0
    
rainfall = [safefloat(row[6]) for row in data]
```

```python
plt.scatter(tmax, rainfall)
```

```python
n = 10
X = np.arange(n)
Y = np.random.normal(0, 100, n)
plt.bar(X, Y)
```

** Problem ** Using the above weather dataset, plot a bar chart of average rain fall per month.

```python
data[:2]
```

```python
months = np.array([row[2] for row in data])
```

```python
rainfall = np.array([float(row[-1] or 0.0) for row in data])
```

```python
months.shape
```

```python
rainfall.shape
```

```python
rainfall[months == 'January'].mean()
```

```python
import datetime
list_of_months = [datetime.date(2000, i+1, 1).strftime("%B") for i in range(12)]
```

```python
list_of_months
```

```python
def get_mean_rainfall(month):
    return rainfall[months == month].mean()
```

```python
mean_rainfall = [get_mean_rainfall(m) for m in list_of_months]
```

```python
mean_rainfall
```

```python
plt.bar(range(12), mean_rainfall)
```
